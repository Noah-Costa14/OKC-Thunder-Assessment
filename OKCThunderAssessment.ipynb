```python
###ASNWERS AT BOTTOM OF FILE###
import pandas as pd
import numpy as np
import math
shotFile = pd.read_csv("shots_data.csv")
shotFile.head()
#shotsFile.tail()
```

```python
#CALCULATING DISTANCE
#SEPARATING SHOTS INTO THE SECTION OF THE FLOOR
#ADDING NEW COLUMNS TO THE ORIGINAL DATAFRAME
#  -TRUE IF SHOT IS IN THE SECTION, FALSE OTHERWISE
distance = (shotFile["x"]**2 + shotFile["y"]**2)**0.5
pt2 = (distance < 23.5) & (shotFile["y"] > 7.8) | (distance < 22) & (shotFile["y"] <= 7.8)
corner3 = (distance >= 22) & (shotFile["y"] <= 7.8)
nonCorner3 = (distance >= 23.5) & (shotFile["y"] > 7.8)

shotFile["pt2"] = pt2
shotFile["corner3"] = corner3
shotFile["nonCorner3"] = nonCorner3
shotFile.head()
```

```python
#FIND THE TOTAL NUMBER OF SHOTS TAKEN BY TEAM A AND TEAM B
#MASKING THE DATAFRAMES TO EACH SECTION OF SHOT
#  -TAKE TOTAL NUMBER OF SHOTS IN EACH SECTION AND DIVIDE BY TOTAL SHOTS TO FIND DISTRIBUTIONS

aTeam = shotFile[shotFile["team"] == "Team A"]
bTeam = shotFile[shotFile["team"] == "Team B"]
aTeamLen = len(aTeam["team"])
bTeamLen = len(bTeam["team"])

aTeam2ptDist = len(aTeam[aTeam["pt2"]] == True) / aTeamLen
bTeam2ptDist = len(bTeam[bTeam["pt2"]] == True) / bTeamLen

aTeamCorner3Dist = len(aTeam[aTeam["corner3"]] == True) / aTeamLen
bTeamCorner3Dist = len(bTeam[bTeam["corner3"]] == True) / bTeamLen

aTeamNC3Dist = len(aTeam[aTeam["nonCorner3"]] == True) / aTeamLen
bTeamNC3Dist = len(bTeam[bTeam["nonCorner3"]] == True) / bTeamLen

print(aTeam2ptDist)
print(aTeamCorner3Dist)
print(aTeamNC3Dist)
print(bTeam2ptDist)
print(bTeamCorner3Dist)
print(bTeamNC3Dist)
```

```python
#CALCULATE FGM, 3PM AND FGA BY EACH TEAM IN EACH SECTION
#CALCULATE eFG%

#eFG = (FGM + (0.5*3PM)) / FGA

a2PM = len(aTeam[(aTeam["fgmade"] == 1) & (aTeam["pt2"] == True)])
aFGA2PM = len(aTeam[aTeam["pt2"] == True])

a3PMCorner = len(aTeam[(aTeam["fgmade"] == 1) & (aTeam["corner3"] == True)])
aFGACorner3 = len(aTeam[aTeam["corner3"] == True])

a3PMNonCorner = len(aTeam[(aTeam["fgmade"] == 1) & (aTeam["nonCorner3"] == True)])
aFGANoneCorner3 = len(aTeam[aTeam["nonCorner3"] == True])

b2PM = len(bTeam[(bTeam["fgmade"] == 1) & (bTeam["pt2"] == True)])
bFGA2PM = len(bTeam[bTeam["pt2"] == True])

b3PMCorner = len(bTeam[(bTeam["fgmade"] == 1) & (bTeam["corner3"] == True)])
bFGACorner3 = len(bTeam[bTeam["corner3"] == True])

b3PMNonCorner = len(bTeam[(bTeam["fgmade"] == 1) & (bTeam["nonCorner3"] == True)])
bFGANonCorner3 = len(bTeam[bTeam["nonCorner3"] == True])


aTeamEFG2PT = (a2PM + (0.5 * 0)) / aFGA2PM
aTeamEFG3PTCorner = (a3PMCorner + (0.5 * a3PMCorner)) / aFGACorner3
aTeamEFG3PTNonCorner = (a3PMNonCorner + (0.5 * a3PMNonCorner)) / aFGANoneCorner3


bTeamEFG2PT = (b2PM + (0.5 * 0)) / bFGA2PM
bTeamEFG3PTCorner = (b3PMCorner + (0.5 * b3PMCorner)) / bFGACorner3
bTeamEFG3PTNonCorner = (b3PMNonCorner + (0.5 * b3PMNonCorner)) / bFGANonCorner3
```

```python
###ANSWERS###
#A TEAM SHOT DISTRIBUTION
#2 Pointer
print("A Team 2 Pointer Shot Distribution: " + str(round(aTeam2ptDist * 100,2)) + "%")

#Corner 3
print("A Team Corner 3 Shot Distribution: " + str(round(aTeamCorner3Dist * 100,2)) + "%")

#Non Corner 3
print("A Team Non-Corner 3 Shot Distribution: " + str(round(aTeamNC3Dist * 100,2)) + "%")
print()

#B TEAM SHOT DISTRIBUTION
#2 Pointer
print("B Team 2 Pointer Shot Distribution: " + str(round(bTeam2ptDist * 100,2)) + "%")

#Corner 3
print("B Team Corner 3 Shot Distribution: " + str(round(bTeamCorner3Dist * 100,2)) + "%")

#Non Corner 3
print("B Team Non-Corner 3 Shot Distribution: " + str(round(bTeamNC3Dist * 100,2)) + "%")
print()

#A TEAM EFFECTIVE FIELD GOAL PERCENTAGE
#2 Pointer
print("A Team Effective Field Goal Percentage (2PT): " + str(round(aTeamEFG2PT * 100,2)) + "%")

#Corner 3
print("A Team Effective Field Goal Percentage (Corner 3): " + str(round(aTeamEFG3PTCorner * 100,2)) + "%")

#Non Corner 3
print("A Team Effective Field Goal Percentage (Non-Corner3): " + str(round(aTeamEFG3PTNonCorner * 100,2)) + "%")
print()

#B TEAM EFFECTIVE FIELD GOAL PERCENTAGE
#2 Pointer
print("B Team Effective Field Goal Percentage (2PT): " + str(round(bTeamEFG2PT * 100,2)) + "%")

#Corner 3
print("B Team Effective Field Goal Percentage (Corner 3): " + str(round(bTeamEFG3PTCorner * 100,2)) + "%")

#Non Corner 3
print("B Team Effective Field Goal Percentage (Non-Corner 3): " + str(round(bTeamEFG3PTNonCorner * 100,2)) + "%")
```
